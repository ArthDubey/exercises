```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
```

```python
spam_data = pd.read_csv("data/spambase.csv")
```

```python
from sklearn.model_selection import train_test_split
spam_train,spam_test = train_test_split(spam_data,test_size=0.2,random_state=100)
features = spam_train.columns.drop("spam")
target = ["spam"]
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=10,
                               min_samples_split=20,
                               min_impurity_decrease=0.05)
model.fit(spam_train[features],spam_train[target])
train_accuracy = model.score(spam_train[features],spam_train[target])
test_accuracy = model.score(spam_test[features],spam_test[target])
print(train_accuracy,test_accuracy)
```

```python
feature_imps = pd.DataFrame(np.array([features,
                                      model.feature_importances_]).T,
                            columns=["feature","importance"])
feature_imps.sort_values(by="importance",ascending=False)
```
